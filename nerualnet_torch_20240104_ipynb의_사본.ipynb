```python
import torch

x_train = torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]])
y_train = torch.FloatTensor([[0],[1],[1],[0]])
W_h = torch.randn([2,3], requires_grad = True)
W_o = torch.randn([3,1], requires_grad = True) # 행렬 크기 왜 그런지 정확히 파악하기.
b_h = torch.randn([3],requires_grad=True)
b_o = torch.randn([1],requires_grad=True)

optimizer = torch.optim.SGD([W_h,W_o,b_h,b_o], lr = 0.01)
#우리가 최적화하고 픈거를 넣어 W랑 b
# deep learning 1단계  : we will make a neural network model

def H(x) :
  HL1 = torch.sigmoid(torch.matmul(x,W_h)+b_h)
  Out = torch.sigmoid(torch.matmul(HL1,W_o)+b_o)
  return Out

# deep learning strp 2 : training.

for step in range(200000):
  cost = torch.mean((-1)*y_train*torch.log(H(x_train))+(-1)*(1-y_train)*torch.log(1-H(x_train)))   # binary에서는 이 return 줄하고 cost 변수부분이 바뀔 것이다.
  # 아래 3줄의 코드는 신경쓰지 않아도 괜찮아.
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
  # print("index and cost: " ,  step, cost )
  print('Step and Cost',step,cost.detach().item()) #오버 슈팅 막으려면..확인해라

#3단계 inference testing

# 만약 성공적이라면 4,4 s는 40around r가 나와야


H_test = H(x_train)
print(H_test)
#print(H(X_train))
```
