```python
import torch

x_train = torch.FloatTensor([[1,2],[2,3],[3,4],[4,4],[5,3],[6,2]])
y_train = torch.FloatTensor([[0],[0],[0],[1],[1],[1]])
W = torch.randn([2,1], requires_grad = True)
b = torch.randn([1],requires_grad=True)
optimizer = torch.optim.SGD([W,b], lr = 0.01)
#우리가 최적화하고 픈거를 넣어 W랑 b
# deep learning 1단계  : we will make a model

def H(x) :
  return torch.sigmoid(torch.matmul(x,W)+b) # x는 1by 2 , W는 2by 1이니까 matrix multiple cation 가능
  # binary에서는 이 return 줄하고 cost 변수부분이 바뀔 것이다.


# deep learning strp 2 : training.

for step in range(2000):
  cost = torch.mean((-1)*y_train*torch.log(H(x_train))+(-1)*(1-y_train)*torch.log(1-H(x_train)))   # binary에서는 이 return 줄하고 cost 변수부분이 바뀔 것이다.
  # 아래 3줄의 코드는 신경쓰지 않아도 괜찮아.
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
  # print("index and cost: " ,  step, cost )
  print('Step and Cost',step,cost.detach().item()) #오버 슈팅 막으려면..확인해라

#3단계 inference testing

# 만약 성공적이라면 4,4 s는 40around r가 나와야

x_test = torch.FloatTensor([[6,1]])
H_test = H(x_test)
print("Testing with [6,1]:" , H_test.detach().item())
```
