```python
import torch

x_train = torch.FloatTensor([[1,2],[2,3],[3,4],[4,4],[5,3],[6,2]])
y_train = torch.FloatTensor([[0],[0],[0],[1],[1],[1]])
W = torch.randn([2,1], requires_grad = True)
b = torch.randn([1],requires_grad=True)
optimizer = torch.optim.SGD([W,b], lr = 0.01)
#    W b
# deep learning 1  : we will make a model

def H(x) :
  return torch.sigmoid(torch.matmul(x,W)+b) # x 1by 2 , W 2by 1 matrix multiple cation 
  # binary  return  cost   .


# deep learning strp 2 : training.

for step in range(2000):
  cost = torch.mean((-1)*y_train*torch.log(H(x_train))+(-1)*(1-y_train)*torch.log(1-H(x_train)))   # binary  return  cost   .
  #  3    .
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
  # print("index and cost: " ,  step, cost )
  print('Step and Cost',step,cost.detach().item()) #  ..

#3 inference testing

#   4,4 s 40around r 

x_test = torch.FloatTensor([[6,1]])
H_test = H(x_test)
print("Testing with [6,1]:" , H_test.detach().item())
```
