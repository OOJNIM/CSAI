```python
import torch

x_train = torch.FloatTesor([[1,1],[2,2],[3,3]])
y_train = torch.FloatTensor([[10],[20],[30]])
W = torch.randn([2,1],requires_grad=True)
b = torch.randn([1],requires_grad=True)
optimizer = torch.optim.SGD([W,b], lr = 0.01)
#우리가 최적화하고 픈거를 넣어 W랑 b
# deep learning 1단계  : we will make a model

def H(x) :
  return torch.matmul(x,W) +b  # x는 1by 2 , W는 2by 1이니까 matrix multiple cation 가능
  # binary에서는 이 return 줄하고 cost 변수부분이 바뀔 것이다.


# deep learning strp 2 : training.

for step in range(2000):
  cost = torch.mean((H(x_train) - y_train)**2)   # binary에서는 이 return 줄하고 cost 변수부분이 바뀔 것이다.
  # 아래 3줄의 코드는 신경쓰지 않아도 괜찮아.
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
  # print("index and cost: " ,  step, cost )
  print('Step and Cost',step,cost.detach().item()) #오버 슈팅 막으려면..확인해라

#3단계 inference testing

# 만약 성공적이라면 4,4 s는 40around r가 나와야

x_test = torch.FloatTensor([[4,4]])
H_test = H(x_test)
print("Testing with [4,4]:" , H_test.detach().item())

#what is the e-05  = 1.7301 multiplication 10 to the -5  , 0.000001732
#according to target performance , 1% accuracy or more , than we need more training.
```

```python
import torch

x_train =torch.FloatTensor([[1,1],[2,2],[3,3]])
y_train = torch.FloatTensor([[10],[20],[30]])
W = torch.randn([2,1],requires_grad = True)
b = torch.randn([1],requires_grad = True)
optimizer = torch.optim.SGD([W,b],lr = 0.01)

def H(x):
	return torch.matmul(x,W) + b

for step in range(2000):
	cost = torch.mean((H(x_train)-y_train)**2)
	optimizer.zero_grad()
	cost.backward()
	optimizer.step()

	print('Step and Cost',step, cost.detach().item())

x_test = torch.FloatTensor([[4,4]])
h_test = H(x_test)
print( 'result of test [4,4]' , h_test.detach().item())
```
