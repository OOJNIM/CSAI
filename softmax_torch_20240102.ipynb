```python
import torch
import torch.nn.functional as F #nn = neural network

x_train = torch.FloatTensor([[1,2,2,2],[1,3,3,3],[1,4,4,4],[1,5,5,5],[1,6,6,6],[1,7,7,7]])
y_train = torch.FloatTensor([[0,0,1],[0,0,1],[0,1,0],[0,1,0],[1,0,0],[1,0,0]])
index_y = torch.argmax(y_train,1)
W = torch.randn([4,3], requires_grad = True)
b = torch.randn([3],requires_grad=True)
optimizer = torch.optim.SGD([W,b], lr = 0.01)


def H(x) :
  return F.softmax(torch.matmul(x,W)+b)

# deep learning strp 2 : training.

for step in range(50000):
  cost = F.cross_entropy(H(x_train),index_y)
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  print('Step and Cost',step,cost.detach().item())

#3단계 inference testing



x_test = torch.FloatTensor([[1,8,8,8]])
H_test = H(x_test)
#print( H_test.item())
print(H_test.detach().item())
```
